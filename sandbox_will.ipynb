```python
import pandas as pd
from pathlib import Path
```

Reading in Jacob's files

```python
DFF = pd.read_csv(Path('Jacob/DFF.csv'), index_col='DATE', parse_dates=True, infer_datetime_format=True)
```

```python
DFF = DFF.groupby(by=[DFF.index.year, DFF.index.month])['DFF'].mean()
```

```python
dff = pd.DataFrame(DFF)
```

```python
dff
```

```python
mortgage_df = pd.read_csv(Path('Jacob/MORTGAGE30US.csv'), index_col='DATE', parse_dates=True, infer_datetime_format=True)
```

```python
mortgage_df
```

```python
temp = mortgage_df.groupby(by=[mortgage_df.index.year, mortgage_df.index.month])['MORTGAGE30US'].mean()
```

```python
mortgage_res = pd.DataFrame(temp)
```

```python
mortgage_res
```

```python
master = pd.concat([dff, mortgage_res], axis=1)
```

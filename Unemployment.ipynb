```python
from pathlib import Path
import pandas as pd
```

```python
MIA = pd.read_csv(Path('MIA_UR.csv'))
BOS = pd.read_csv(Path('BOS_UR.csv'))
PHO = pd.read_csv(Path('PHO_UR.csv'))
NYC = pd.read_csv(Path('NYC_UR.csv'))
SEA = pd.read_csv(Path('SEA_UR.csv'))
LA = pd.read_csv(Path('LA_UR.csv'))
HOU = pd.read_csv(Path('HOU_UR.csv'))
CHI = pd.read_csv(Path('CHI_UR.csv'))

df_list=[MIA, BOS, PHO, NYC, SEA, LA, HOU, CHI]
```

```python
df_master = pd.concat(df_list, axis=1)
```

```python
import pandas as pd
from pathlib import Path
```

```python
MIA = pd.read_csv(Path('Resources_Will/MIA_Crime.csv'))
BOS = pd.read_csv(Path('Resources_Will/BOS_Crime.csv'))
PHO = pd.read_csv(Path('Resources_Will/PHO_Crime.csv'))
NYC = pd.read_csv(Path('Resources_Will/NYC_Crime.csv'))
SEA = pd.read_csv(Path('Resources_Will/SEA_Crime.csv'))
LA = pd.read_csv(Path('Resources_Will/LA_Crime.csv'))
HOU = pd.read_csv(Path('Resources_Will/HOU_Crime.csv'))
CHI = pd.read_csv(Path('Resources_Will/CHI_Crime.csv'))

df_list=[MIA, BOS, PHO, NYC, SEA, LA, HOU, CHI]
```

```python
MIA['series'] = 'MIA'
BOS['series'] = 'BOS'
PHO['series'] = 'PHO'
NYC['series'] = 'NYC'
SEA['series'] = 'SEA'
LA['series'] = 'LA'
HOU['series'] = 'HOU'
CHI['series'] = 'CHI'
```

```python
remove_cleared = []
for df in df_list:
    df = df.drop(1)
    remove_cleared.append(df)
```

```python
master = pd.concat(remove_cleared, axis=0)
master_crime = master.set_index('series').T
```

```python
master_crime = master_crime.rename(columns={'MIA':'MIA_CRIME',
                         'BOS':'BOS_CRIME',
                         'PHO': 'PHO_CRIME',
                         'NYC':'NY_CRIME',
                         'SEA':'SEA_CRIME',
                         'LA':'LA_CRIME',
                         'HOU':'HOU_CRIME',
                         'CHI':'CHI_CRIME'
})
```

```python
master_crime.reset_index(inplace=True)
```

```python
master_crime
```

```python
master_crime['index']=pd.to_datetime(master_crime['index'])
master_crime = master_crime.set_index('index').resample('M').last().fillna(method='ffill').reset_index()
```
